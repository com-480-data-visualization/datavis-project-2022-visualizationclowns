```python
import pandas as pd
```

```python
tweets = pd.read_csv("../tweets/alltweets.csv", index_col=0)
tweets.created_at = pd.to_datetime(tweets.created_at)
tweets = tweets[::-1].reset_index(drop=True)
tweets
```

```python
mentions = tweets.tweet.str.extractall(r'(@\w+)').groupby(level=0).agg(' '.join)
mentions
```

```python
links = tweets.tweet.str.extractall(r'(http://t.co/\w+)').groupby(level=0).agg(' '.join)
links
```

```python
hashtags = tweets.tweet.str.extractall(r'(#\w+)').groupby(level=0).agg(' '.join)
hashtags
```

```python
clean_tweets = tweets.tweet.str.replace(r'http\S+|#\w+|@\w+', '', regex=True)
clean_tweets
```

```python
new_tweets = tweets.copy()
new_tweets["unclean_tweet"] = new_tweets["tweet"]
new_tweets["mentions"] = mentions
new_tweets["tweet"] = clean_tweets
new_tweets["hashtags"] = hashtags
new_tweets["links"] = links
new_tweets
```

```python
new_tweets.to_csv("../tweets/alltweets_clean.csv")
```
