We have a totoal of 15583 Tweets between the 2022-03-07 07:49:42 and 2010-06-04 18:31:57. 

The main data we have available for the data exploration is the date of the tweet, its contents as well as the number of likes, replies and retweets.

```python
tweets = pd.read_csv("../tweets/alltweets.csv")
tweets.shape
```

We have 50 tweets that cointain the work bitcoin or BTC. This seems as not that much but we will se what we can make of it

```python
bitcoin_tweets = tweets[tweets["tweet"].str.match(r".*([b,B]itcoin|BTC).*")].reset_index()
shape = bitcoin_tweets.shape
bitcoin_tweets.head(shape[0])
```

We have 120 tweets with Doge in the tweets.

```python
dogecoin_tweets = tweets[tweets["tweet"].str.match(r".*((D|d)(O|o)(G|g)(E|e)).*")].reset_index()
dogecoin_tweets.head(dogecoin_tweets.shape[0])
```

We have 2988 tweets with TSLA or Tesla in the tweets.

```python
tsla_tweets = tweets[tweets["tweet"].str.match(r".*((T|t)(S|es)(L|l)(A|a)).*")].reset_index()
tsla_tweets.head(tsla_tweets.shape[0])
```
